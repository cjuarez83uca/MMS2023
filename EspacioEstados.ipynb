```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig
import control as ct
```

```python
# wn  Z
wn = 2 * np.pi * 10
Z = 0.8

# constantes del sistema
m = 2.5
k = wn**2*m
c = 2*Z*wn*m
```

```python
# representacion en espacio de estado
A=np.array([[0, 1],[-k/m, -c/m]])
B=np.array([[0],[1/m]])
C=np.array([[0, 1],[1, 0],[-k/m, -c/m]])
D=np.array([[0], [0], [1/m]])
```

```python
l=eig(A)
l
```

```python
sys = ct.ss(A,B,C,D)
sys
```

```python
t= ct.step_response(sys,0.2)
```

```python
plt.plot(t.time,t.outputs[0][0])
```
