```python
# Problema de Sistema masa resorte amortiguador
# importar librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Constantes del sistema
M = 1 # kg
c = 8 # kg/s
k = 1 # kg/s^2
```

```python
t = np.arange(0,10,0.01)
```

```python
# condiciones iniciales
x0 = 1
v0 = 0
```

```python
p1 = -c/(2*M)+(1*pow(c/(2*M),2)-k/M)**0.5
print(p1)
```

```python
p2 = -c/(2*M)-(1*pow(c/(2*M),2)-k/M)**0.5
print(p2)
```

```python
A1 = (x0*p2-v0)/(p2-p1)
print(A1)
```

```python
A2 = (v0-p1*x0)/(p2-p1)
print(A2)
```

```python
x = A1*np.exp(p1*t)+A2*np.exp(p2*t)

plt.plot(t,x)
```
