```python
# euler

tfin = 16
h = 1
n = int(tfin/1 + 1)

# vectores de tiempo y velocidad con ceros.
t = np.zeros(n)
v = np.zeros(n)

# EDO Condiciones iniciales
v[0] = 0 # parte del reposo


# m√©todo de euler

for i in np.arange(1,n):
    t[i] = t[i-1] + h
    v[i] = v[i-1] + h*f(t[i-1],v[i-1])


plt.plot(t,v)

```
