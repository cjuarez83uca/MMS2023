```python
c = 0.5
m = 2.5

def f(t,v):
    if t >= 1:
        F = 1
    else:
        F = 0
    sal = -c / m * v + 1 / m * F
    return sal
```

```python
# euler
# vectores
tfin = 15
h = 1
n = int(tfin/h + 1)
t = np.zeros(n)
v = np.zeros(n)

# condiciones iniciales
v[0] = 0

for i in np.arange(1,n):
    t[i] = t[i-1] + h
    v[i] = v[i-1] + h * f(t[i-1],v[i-1])
```

```python
plt.plot(t,v)
```
