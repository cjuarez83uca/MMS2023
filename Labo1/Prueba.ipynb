```python
c = 0.8
m = 2.5
# creacion de funcion
def f(t,v):
    
    #if t >= 5:
     #   F = 1
    #else:
     #   F = 0
    F = 0.1*np.sin(2*np.pi*t*1)
    sal = -c/m*v + 1/m*F
    return sal

# variable velocidad y tiempo
dt = 0.05
tfin = 15
n = int(tfin/dt + 1)
n

t = np.zeros(n)
v = np.zeros(n)

for i in np.arange(1,n):
    t[i] = t[i-1] + dt
    v[i] = v[i-1] + dt * f(t[i-1],v[i-1])

plt.plot(t,v)
```
