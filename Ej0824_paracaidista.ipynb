```python
# Programa del paracaidista(otra vez)
# el numeral al inicio de la línea implica comentario
# todo lo que este a la derecha se ignora 

# Primer paso importar librerias que utilizaré en mi código
import numpy as np
import matplotlib.pyplot as plt

# jugando con numpy
a = np.arange(60)
#print(a)
#print('Dimensiones de a es:  ',np.size(a))

# Constantes del problema 
g = 9.81 # Gravedad m/s2
c = 12.5 # Coeficiente de arrastre kg/s
m = 68.1 # Masa de paracaidista kg

def f(v):
    sal = g - c/m * v
    
    return sal

# Definiendo datos de tiempo para la simulacion
h = 1 # paso de integracion, delta de tiempo
tfin = 10 # tiempo final de simulacion
n = int(tfin/h+1) # cantidad de datos de tiempo y velocidad
print(n)
# inicializando variables del programa parte iterativa
t = np.zeros(n)
v = np.zeros(n)

print(t)
```

```python
# resolver la ecuacion diferencial por medio de Euler
# Condiciones iniciales
v[0] = 0 # parte del reposo

for i in np.arange(1,n): # i contador de iteraciones, np.arange(1,n) valores q se mueve i
    v[i] = v[i-1] + h * f(v[i-1])
    t[i] = t[i-1] + h
print(v)
print(t)

# solucion analitica
v_real = g*m/c*(1-np.exp(-c/m*t))

plt.plot(t,v_real,'b',t,v,'r')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad m/s')
plt.title('Grafica de paracaidista')
plt.legend(['Real','Euler'])
```
