```python
# importar librerias que necesito
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constantes
D = 0.1 #kg/s
m = 12.5 #kg
v0 = 5 # m/s
```

```python
t=np.arange(0,5*125,2)
v = v0*np.exp(-D/m*t)
```

```python
plt.plot(t,v)
```

```python
# solucion numerica Euler
def f(t,v):
    res=-D/m*v
    return res
```

```python
# discretizacion de variable tiempo
h = 2
tsimu = 5*125;
n = int(tsimu/h+1)
```

```python
# variables computacionales
v_e = np.zeros(n) # velocidad por euler
t_e = np.zeros(n) # tiempo en euler
```

```python
# condicion inicial
v_e[0] = v0
```

```python
# aplicando euler
for i in np.arange(1,n):
    v_e[i] = v_e[i-1] + h * f(t_e[i-1],v_e[i-1])
    t_e[i] = t_e[i-1] + h
```

```python
plt.plot(t,v,t_e,v_e)
```
